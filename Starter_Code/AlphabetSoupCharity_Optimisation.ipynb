```python



# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
#  Import and read the charity_data.csv.
from google.colab import files
uploaded = files.upload()
import os 
import tensorflow
from tensorflow import keras
```

```python
#  Import and read the charity_data.csv.
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' .
application_df = application_df.drop(columns = ['EIN'])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at IS_SUCCESSFUL  value counts for binning
counts = application_df['NAME'].value_counts()
counts

     
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
names_to_replace = list(counts[counts<5].index)
names_to_replace

# Replace in dataframe
for names in names_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(names,"Other")

# Check to make sure binning was successful
application_df['NAME'].value_counts()



```

```python
# Look at APPLICATION TYPE counts for binning
counts = application_df['APPLICATION_TYPE'].value_counts()
counts



#countsBinning = application_df['CLASSIFICATION'].value_counts()
#classificationValueCounts = countsBinning[countsBinning>1]
#classificationValueCounts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(counts[counts<500].index)
application_types_to_replace

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
countsbinning = application_df['CLASSIFICATION'].value_counts()


     
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
countsclassification = countsbinning[countsbinning>1]
countsclassification
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list (countsbinning[countsbinning<100].index)
classifications_to_replace
```

```python
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()

# Look at CLASSIFICATION value counts for binning
#countsbinning = application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# application_df = pd.get_dummies(application_df,dtype=float)
application_df = pd.get_dummies(application_df,dtype=float)
application_df.head()




```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL'].values
y



# Split the preprocessed data into a training and testing dataset
X = application_df.drop('IS_SUCCESSFUL', axis=1).values
X

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()



# Fit the StandardScaler
X_scaler = scaler.fit(X_train)



# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  # Define the model - deep neural net

# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len( X_train_scaled[0])
hidden_nodes_layer1=6
hidden_nodes_layer2=13
hidden_nodes_layer3=16
nn = tf.keras.models.Sequential()



# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation='tanh'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
     





```

```python
#Define checkpoints and filenames
os.makedirs("checkpoint/",exist_ok=True)
checkpoint_path="checkponts/weights.{epoch:02d}.hdfs"
```

```python
from keras.callbacks import ModelCheckpoint
#Create a callback that saves the model's weight every 5 epoch
cp_callback=ModelCheckpoint(filepath=checkpoint_path,verbose=1,save_weights_only=True,save_freq=5)
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

```

```python
# Train the model
#fit_model = nn.fit(X_train_scaled,y_train,epochs=50)

fit_model = nn.fit(X_train_scaled,y_train,validation_split=0.15,epochs=100,callbacks=[cp_callback])
#tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
# Train the model
#fit_model = nn.fit(X_train_scaled,y_train,validation_split=0.15, epochs=100)


```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
